### Objective:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## 📂 Data Loading

```python
df = pd.read_csv("superstore_cleaned.csv")
df.head()
```

## 🔍 Data Overview
Understanding the structure and basic statistics of the dataset.

```python
df.shape
df.info()
df.describe()
```

```python
sns.histplot(df['Sales'], bins=30)
plt.title("Sales Distribution")
plt.show()
```

```python
sns.histplot(df['Profit'], bins=30)
plt.title("Profit Distribution")
plt.show()
```

```python
sns.countplot(x='Category', data=df)
plt.title("Category Distribution")
plt.show()
```

```python
sns.countplot(x='Segment', data=df)
plt.title("Customer Segments")
plt.show()
```

## 📌 Business Questions & Analysis
Answering key business questions using data.

### Top 5 States by Sales

```python
df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(5)
```

### Top 5 Cities by Profit

```python
df.groupby('City')['Profit'].sum().sort_values(ascending=False).head(5)
```

### Sales by Category

```python
df.groupby('Category')['Sales'].sum()
```

### Profit by Sub-Category

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values()
```

### Region-wise Sales

```python
df.groupby('Region')['Sales'].sum()
```

### Impact of Discount on Profit

```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

## 📊 Multivariate Analysis
Exploring relationships between multiple variables.

### Correlation Heatmap

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

### Sales vs Profit

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.show()
```

### Profit by Category

```python
sns.boxplot(x='Category', y='Profit', data=df)
plt.title("Profit by Category")
plt.show()
```

### Sales by Region

```python
sns.barplot(x='Region', y='Sales', data=df)
plt.title("Sales by Region")
plt.show()
```

## 📈 Key Performance Indicators (KPIs)

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
avg_discount = df['Discount'].mean()

top_category = df.groupby('Category')['Sales'].sum().idxmax()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
print("Average Discount:", avg_discount)
print("Top Category:", top_category)
```

## 🧠 Key Insights

- The West region generates the highest sales.
- Technology category contributes the most to overall profit.
- Higher discounts tend to reduce profitability.
- Certain sub-categories like Tables show consistent losses.
- Sales and profit are not always directly proportional.
- A few cities dominate both sales and profit performance.

## ✅ Conclusion

This analysis provided valuable insights into sales performance, profitability, and customer behavior. 
The findings can help businesses optimize pricing strategies, focus on high-performing categories, 
and reduce losses from underperforming segments.
